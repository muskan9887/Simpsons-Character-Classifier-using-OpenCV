### Image Transformations

```python
img = cv2.imread("/content/cat.jpeg")
reimg = cv2.resize(img, (250,250))
cv2_imshow(reimg)
```

```python
def translate(reimg, x, y):
  transMat = np.float32([[1,0,x],[0,1,y]])
  dimensions = (reimg.shape[1], reimg.shape[0])
  return cv2.warpAffine(reimg, transMat, dimensions)

# -x-->left
# -y-->up
# x-->right
# y-->down

translated = translate(reimg, -100, 100)
cv2_imshow(translated)
```

```python
#Rotation

def rotate(reimg, angle, rotPoint=None):
  (height, width) = reimg.shape[:2]

  if rotPoint is None:
    rotPoint = (width//2, height//2)

  rotMat = cv2.getRotationMatrix2D(rotPoint, angle, 1.0)
  dimensions = (width, height)

  return cv2.warpAffine(reimg, rotMat, dimensions)

rotated = rotate(reimg, 45)

cv2_imshow(rotated)
```

```python
#Flipping

flip = cv2.flip(reimg, -1)
cv2_imshow(flip)
```

#### Bitwise operators in opencv

```python
blank = np.zeros((200,200),dtype='uint8')
rectangle = cv2.rectangle(blank.copy(), (30,30), (170,170), 70, -1)
circle = cv2.circle(blank.copy(), (150,150),75, 255, -1)
cv2_imshow(rectangle)
cv2_imshow(circle)
```

```python
#bitwise and

bitwise_and = cv2.bitwise_and(rectangle, circle)
cv2_imshow(bitwise_and)
```

```python
#bitwise or

bitwise_or = cv2.bitwise_or(rectangle, circle)
cv2_imshow(bitwise_or)
```

```python
#bitwisexor

bitwise_xor = cv2.bitwise_xor(rectangle, circle)
cv2_imshow(bitwise_xor)
```

```python
#bitwise not

bitwise_nor = cv2.bitwise_not(rectangle,circle)
cv2_imshow(bitwise_nor)
```
